# Creating a single function in which the piecing method can be used:

### Creating test data $\rightarrow$ ombh2 = ombh2(Planck2018)*1.2

```python
pars_test = camb.set_params(H0=67.5, 
                            ombh2=0.022, 
                            omch2=0.120*1.2, 
                            redshifts=[0], 
                            kmax=10, 
                            As=2e-9, 
                            ns=0.96, 
                            halofit_version='mead2020_feedback', 
                            HMCode_A_baryon = 0, 
                            HMCode_eta_baryon = 0, 
                            HMCode_logT_AGN = -10)
                            # zero the HMCode to get a universe without baryonic feedback
results_test = camb.get_transfer_functions(pars_test)
kh_test, _, pk_test = results_test.get_nonlinear_matter_power_spectrum(hubble_units=True, k_hunit=True)

kh_data = kh_test
pk_data = pk_test[0]
```

### Creating cosmology using Planck2018

```python
pars_planck_2018 = camb.set_params(H0=67.4, ombh2=0.0224, omch2=0.120,
                                  redshifts=[0], kmax=10, As=2e-9, ns=0.96, 
                                  halofit_version='mead2020_feedback',
                                  HMCode_A_baryon=0, HMCode_eta_baryon=0, HMCode_logT_AGN=-10)
results_planck_2018 = camb.get_transfer_functions(pars_planck_2018)
kh_planck2018, _, pk_planck2018 = results_planck_2018.get_nonlinear_matter_power_spectrum(hubble_units=True, k_hunit=True)
```

```python
def baryon_model(khdata, pkdata):
    
    def baryon_model_mead(khdata, A_baryon, eta_baryon, logT_AGN):
        pars_mead = camb.set_params(H0=67.5, ombh2=0.022, omch2=0.122, redshifts=[0], kmax=10, As=2e-9, ns=0.96, 
                           halofit_version='mead', 
                           HMCode_A_baryon = A_baryon, 
                           HMCode_eta_baryon = eta_baryon, 
                           HMCode_logT_AGN = logT_AGN)
        results_mead = camb.get_transfer_functions(pars_mead)
        kh_mead, _, pk_mead = results_mead.get_nonlinear_matter_power_spectrum(hubble_units=True, k_hunit=True)
        return pk_mead[0]
    
    guesses = np.array([3.13, 0.603, 7.8])
    popt_mead, pcov_mead = scipy.optimize.curve_fit(baryon_model_mead, xdata=khdata, ydata=pkdata, p0=guesses) 
    A_best_fit = popt_mead[0]
    eta_best_fit = popt_mead[1]
    
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
    
    def baryon_model_feedback(khdata, A_baryon, eta_baryon, logT_AGN):
        pars_feedback = camb.set_params(H0=67.5, ombh2=0.022, omch2=0.122, redshifts=[0], kmax=10, As=2e-9, ns=0.96, 
                               halofit_version='mead2020_feedback', 
                               HMCode_A_baryon = A_baryon, 
                               HMCode_eta_baryon = eta_baryon, 
                               HMCode_logT_AGN = logT_AGN)
        results_feedback = camb.get_transfer_functions(pars_feedback)
        kh_feedback, _ , pk_feedback = results_feedback.get_nonlinear_matter_power_spectrum(hubble_units=True, k_hunit=True)
        return pk_feedback[0]
    
    popt_feedback, pcov_feedback = scipy.optimize.curve_fit(baryon_model_feedback, xdata=khdata, ydata=pkdata, p0=guesses)
    logT_AGN_best_fit = popt_feedback[2]
    
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
    
    pieced_best_fit_params = np.array([A_best_fit, eta_best_fit, logT_AGN_best_fit])
    
    pars_best_fit = camb.set_params(H0=67.5, ombh2=0.022, omch2=0.122, redshifts=[0], kmax=10, As=2e-9, ns=0.96, 
                               halofit_version='mead2020_feedback', 
                               HMCode_A_baryon = A_best_fit, 
                               HMCode_eta_baryon = eta_best_fit, 
                               HMCode_logT_AGN = logT_AGN_best_fit)
    results_best_fit = camb.get_transfer_functions(pars_best_fit)
    kh_best_fit, _, pk_best_fit = results_best_fit.get_nonlinear_matter_power_spectrum(hubble_units=True, k_hunit=True)
    
    return pieced_best_fit_params, pk_best_fit[0]
```

```python
best_fit_params = baryon_model(kh_data, pk_data)[0]
pk_best_fit_curve = baryon_model(kh_data, pk_data)[1]
```

```python
plt.figure()
fig, ax0 = plt.subplots(ncols=1, nrows=1, 
                               figsize=(8,6), 
                               constrained_layout=True, 
                               sharex=True)
fig.suptitle('Testing Best Fit Code (mead2020_feedback vs. piecing)')

ax0.loglog(kh_data, pk_data, 'k-', label='Observation data [ombh2*1.2]')
ax0.loglog(kh_planck2018, pk_planck2018[0], 'g-', label='Planck2018 cosmology,\nno baryonic feedback')
ax0.loglog(kh_data, pk_best_fit_curve)
#ax0.loglog(kh_data, pk_feedback, 'b-', label='Best fit (mead2020_feedback only)')
#ax0.loglog(kh_data, pieced_together_best_fit, 'r-', label='Best fit (pieced together)')
ax0.set_ylabel('$P_m(k)$ [($h^{-1}Mpc)^3$]')
ax0.legend(loc='lower left')
```
