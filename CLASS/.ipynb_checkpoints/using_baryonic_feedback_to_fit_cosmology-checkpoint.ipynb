```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize
from classy import Class


%matplotlib inline
%config InlineBackend.figure_format = 'retina'
plt.rcParams['mathtext.fontset'], plt.rcParams['font.family'] = 'stix', 'STIXGeneral'
plt.rcParams.update({'font.size': 17})
```

### Create "fake" observational data

Changing omch2$\rightarrow$omch2*1.2 (off-set cosmology from Planck2018) and zeroing baryonic feedback (for universe without feedback)

```python
omch2 = 0.120*1.2 # changing amount of CDM in universe by 20%

data = Class() 

data.set({'omega_cdm':omch2,
          'output':'mPk',
          'P_k_max_h/Mpc':10, 
          'lensing':'no', 
          'z_pk':0, 
          }) # not specifying non-linear model --> linear == no baryonic feedback 

data.compute() 
```

```python
ks_data = np.logspace(-5, 1, 1000) # k in h/Mpc

Pks_data = [] 
h = data.h() # get reduced Hubble for conversions to 1/Mpc
for k in ks_data:
    Pk = data.pk(k*h, 0)*h**3 # function .pk(k,z)
    Pks_data.append(Pk) 

Pks_data = np.array(Pks_data) # P(k) in (Mpc/h)**3
```

### Defaults for reference

Planck2018 cosmology and baryonic feedback ('owls_agn' default params)

```python
default = Class() 

default.set({'output':'mPk',
          'non_linear':'hmcode', 
          'feedback model':'owls_agn', 
          'P_k_max_h/Mpc':10, 
          'lensing':'no', 
          'z_pk':0
          })

default.compute()
```

```python
ks_default = np.logspace(-5, 1, 1000) # k in h/Mpc

Pks_default = [] 
h = default.h() # get reduced Hubble for conversions to 1/Mpc
for k in ks_default:
    Pk = default.pk(k*h, 0)*h**3 # function .pk(k,z)
    Pks_default.append(Pk) 

Pks_default = np.array(Pks_default) # P(k) in (Mpc/h)**3
```

### Quick Plot

```python
plt.figure()
plt.loglog(ks_data, Pks_data, 'r', label="obs. data [omch2*1.2]")
plt.loglog(ks_default, Pks_default, 'k', label="defaults")
plt.legend()
plt.show()
```

# Attempting to best fit using baryonic feedback

## scipy.optimize.curve_fit()

```python
def baryon_model(x_data, eta, c):
    model = Class()
    model.set({'output':'mPk',
               'non_linear':'hmcode', 
               'eta_0':eta,
               'c_min':c,
               'P_k_max_h/Mpc':10, 
               'lensing':'no', 
               'z_pk':0
               })
    model.compute()
    
    ks_model = x_data

    Pks_model = [] 
    h = model.h() 
    for k in ks_model:
        Pk = model.pk(k*h, 0)*h**3
        Pks_model.append(Pk) 

    Pks_model = np.array(Pks_model) # P(k) in (Mpc/h)**3
    
    return Pks_model
```

```python
guesses = np.array([1, 1])
popt, pcov = scipy.optimize.curve_fit(baryon_model, ks_data, Pks_data, guesses)
popt
```

## scipy.optimize.least_squares()

```python
def baryon_model_ls(baryons):
    model = Class()
    model.set({'output':'mPk',
          'non_linear':'hmcode',
          'eta_0':baryons[0],
          'c_min':baryons[1], 
          'P_k_max_h/Mpc':10, 
          'lensing':'no', 
          'z_pk':0, 
          })
    model.compute()
    
    ks_model = np.logspace(-5, 1, 1000)
    Pks_model = [] 
    h = model.h() 
    for k in ks_model:
        Pk = model.pk(k*h, 0)*h**3
        Pks_model.append(Pk) 
    Pks_model = np.array(Pks_model)
    
    return ks_model, Pks_model
```

```python
def residuals(baryons, y_data, x_data):
    return y_data - baryon_model_ls(baryons)[1]
```

```python
guesses = np.array([1, 1])

ls = scipy.optimize.least_squares(residuals, guesses, args=(Pks_data, ks_data))
ls.x
```

NOTE: scipy.optimize.least_squares() function took significantly less time to run.

## Plotting and comparing results

```python
plt.figure()
fig, (ax0, ax1) = plt.subplots(ncols=1, nrows=2, 
                               figsize=(12,8), 
                               constrained_layout=True, 
                               sharex=True, 
                               gridspec_kw={'height_ratios':[2, 1]})

ax0.loglog(ks_data, Pks_data, 'green', label='Observational data [omch2*1.2]')
ax0.loglog(ks_default, Pks_default, 'purple', linestyle=':', label='CLASS defaults (reference)')
ax0.loglog(baryon_model_ls(popt)[0], baryon_model_ls(popt)[1], 'b', linestyle='--', label='scipy.optimize.curve_fit()\n[{0:.4f}, {1:.4f}]'.format(*popt))
ax0.loglog(baryon_model_ls(ls.x)[0], baryon_model_ls(ls.x)[1], 'red', linestyle='-.', label='scipy.optimize.least_squares()\n[{0:.4f}, {1:.4f}]'.format(*ls.x))

ax0.grid()
ax0.set_title('Attemping to use baryonic feedback to fit different cosmology using CLASS')
ax0.set_ylabel('$P_m(k)$ [($h^{-1}Mpc)^3$]')
ax0.legend(bbox_to_anchor = (1, 1), loc='upper left')

ax1.semilogx(baryon_model_ls(popt)[0], baryon_model_ls(popt)[1]/Pks_data, color='blue', linestyle='--', label='curve_fit()/obs.')
ax1.semilogx(baryon_model_ls(ls.x)[0], baryon_model_ls(ls.x)[1]/Pks_data, color='red', linestyle='-.', label='least_squares()/obs.')
ax1.legend(bbox_to_anchor = (1, 1), loc='upper left')
ax1.axhline(1, color='grey')
ax1.set_ylabel('$P_{best}/P_{obs.}$', fontsize=19)
ax1.set_xlabel('Wavenumber, k [h/Mpc]')
ax1.grid()
plt.show()
```
