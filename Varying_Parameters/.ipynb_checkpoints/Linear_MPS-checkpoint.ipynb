```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import sys, platform, os
import matplotlib as mpl
from matplotlib import pyplot as plt
import numpy as np
import mpl_toolkits.axes_grid1 as axgrid
from mpl_toolkits.axes_grid1 import make_axes_locatable
plt.rcParams['mathtext.fontset'], plt.rcParams['font.family'] = 'stix', 'STIXGeneral'
plt.rcParams.update({'font.size': 15})

camb_path = os.path.realpath(os.path.join(os.getcwd(),'..'))
sys.path.insert(0,camb_path)
import camb
from camb import model, initialpower
print('Using CAMB %s installed at %s'%(camb.__version__, os.path.dirname(camb.__file__)))
```

# Varying H0

```python
def pars_hubble(h0):
    kh_list = []
    pk_list =[]
    for i in h0:
        pars = camb.CAMBparams()
        pars.set_cosmology(H0=i, ombh2=0.022, omch2=0.120)
        pars.InitPower.set_params(ns=0.965)
        pars.set_matter_power(redshifts=[0], kmax=2.0)

        pars.NonLinear = model.NonLinear_none
        results = camb.get_results(pars)
        kh, z, pk = results.get_matter_power_spectrum(minkh=1e-4, maxkh=1, npoints = 200)
        kh_list.append(kh), pk_list.append(pk[0])
    return kh_list, pk_list
```

```python
Hubble_range = np.arange(63, 80, 1)

cmap = plt.cm.coolwarm
n = len(Hubble_range)
colors = cmap(np.linspace(0,1,n))

fig, ax = plt.subplots(1, 1, figsize = (8,6))
for i in range(len(pars_hubble(Hubble_range)[1])):
    ax.loglog(pars_hubble(Hubble_range)[0][i],
              pars_hubble(Hubble_range)[1][i], color=colors[i])
    
norm_scaling = mpl.colors.Normalize(vmin=min(Hubble_range), vmax=max(Hubble_range))
divider = axgrid.make_axes_locatable(ax)
cax = divider.append_axes("right", size='5%', pad=0.05)
cbar = plt.colorbar(mpl.cm.ScalarMappable(norm=norm_scaling, cmap=cmap), cax=cax)
cbar.set_label("Hubble Constant, $H_o$ [km/s/Mpc]", rotation=270, labelpad=30)

ax.set_xlabel('Wavenumber, k [h/Mpc]')
ax.set_ylabel('$P_m(k)$ [($h^{-1}Mpc)^3$]')
ax.set_title('Matter Power Spectra for Varying Hubble Constants')
plt.show()
```

# Varying ombh2

```python
def pars_ombh2(ombh2):
    kh_list = []
    pk_list =[]
    for i in ombh2:
        pars = camb.CAMBparams()
        pars.set_cosmology(H0=67.4, ombh2=i, omch2=0.120)
        pars.InitPower.set_params(ns=0.965)
        pars.set_matter_power(redshifts=[0], kmax=2.0)
        
        pars.NonLinear = model.NonLinear_none
        results = camb.get_results(pars)
        kh, z, pk = results.get_matter_power_spectrum(minkh=1e-4, maxkh=1, npoints = 200)
        kh_list.append(kh), pk_list.append(pk[0])
    return kh_list, pk_list
```

```python
ombh2_range = np.arange(0.018, 0.026, 0.0005)

cmap = plt.cm.coolwarm
n = len(ombh2_range)
colors = cmap(np.linspace(0,1,n))

fig, ax = plt.subplots(1, 1, figsize = (8,6))
for i in range(len(pars_ombh2(ombh2_range)[1])):
    ax.loglog(pars_ombh2(ombh2_range)[0][i],
              pars_ombh2(ombh2_range)[1][i], color=colors[i])

norm_scaling = mpl.colors.Normalize(vmin=min(ombh2_range), vmax=max(ombh2_range))
divider = axgrid.make_axes_locatable(ax)
cax = divider.append_axes("right", size='5%', pad=0.05)
cbar = plt.colorbar(mpl.cm.ScalarMappable(norm=norm_scaling, cmap=cmap), cax=cax)
cbar.set_label("Baryonic Density, $\Omega_bh^2$", rotation=270, labelpad=30)

ax.set_xlabel('Wavenumber, k [h/Mpc]')
ax.set_ylabel('$P_m(k)$ [($h^{-1}Mpc)^3$]')
ax.set_title('Matter Power Spectra for Varying Baryonic Densities')
plt.show()
```

# Varying omch2

```python
def pars_omch2(omch2):
    kh_list = []
    pk_list =[]
    for i in omch2:
        pars = camb.set_params(H0=67.4, ombh2=0.0224, omch2=i, redshifts=[0], kmax=10, As=2e-9, ns=0.96)

        results = camb.get_transfer_functions(pars)
        kh, z, pk = results.get_linear_matter_power_spectrum(hubble_units=True, k_hunit=True)
        kh_list.append(kh), pk_list.append(pk[0])
    return kh_list, pk_list
```

```python
omch2_range = np.arange(0.11, 0.20, 0.005)

cmap = plt.cm.coolwarm
n = len(omch2_range)
colors = cmap(np.linspace(0,1,n))

fig, ax = plt.subplots(1, 1, figsize = (8,6))
for i in range(len(pars_omch2(omch2_range)[1])):
    plt.loglog(pars_omch2(omch2_range)[0][i],
              pars_omch2(omch2_range)[1][i], color=colors[i])

norm_scaling = mpl.colors.Normalize(vmin=min(omch2_range), vmax=max(omch2_range))
divider = axgrid.make_axes_locatable(ax)
cax = divider.append_axes("right", size='5%', pad=0.05)
cbar = plt.colorbar(mpl.cm.ScalarMappable(norm=norm_scaling, cmap=cmap), cax=cax)
cbar.set_label("CDM Density, $\Omega_ch^2$", rotation=270, labelpad=30)

ax.set_xlabel('Wavenumber, k [h/Mpc]')
ax.set_ylabel('$P_m(k)$ [($h^{-1}Mpc)^3$]')
ax.set_title('Matter Power Spectra for Varying CDM Densities')
plt.show()
```
