```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import sys, platform, os
import matplotlib as mpl
from matplotlib import pyplot as plt
import numpy as np
import mpl_toolkits.axes_grid1 as axgrid
from mpl_toolkits.axes_grid1 import make_axes_locatable
plt.rcParams['mathtext.fontset'], plt.rcParams['font.family'] = 'stix', 'STIXGeneral'
plt.rcParams.update({'font.size': 15})

camb_path = os.path.realpath(os.path.join(os.getcwd(),'..'))
sys.path.insert(0,camb_path)
import camb
from camb import model, initialpower
print('Using CAMB %s installed at %s'%(camb.__version__, os.path.dirname(camb.__file__)))
```

# Varying A_baryon

```python
def varying_A_baryon(A):
    kh_nonlin_list = []
    pk_nonlin_list = []
    
    for i in A:
        pars = camb.read_ini(os.path.join(camb_path, 'Jia Liu', 'planck_2018.ini.txt'))
        pars.set_matter_power(redshifts=[0.], kmax=2.0)
        results = camb.get_results(pars)
        results.Params.NonLinearModel.set_params(halofit_version = 'mead', HMCode_A_baryon = i)
        results.calc_power_spectra()
        kh_nonlin, _, pk_nonlin = results.get_nonlinear_matter_power_spectrum()
        kh_nonlin_list.append(kh_nonlin), pk_nonlin_list.append(pk_nonlin[0])
        
    return kh_nonlin_list, pk_nonlin_list
```

```python
A_baryon_range = np.arange(2.43, 3.83, 0.1)

cmap = plt.cm.RdYlGn
n = len(A_baryon_range)
colors = cmap(np.linspace(0,1,n))

fig, ax = plt.subplots(1, 1, figsize = (8,6))
for i in range(len(varying_A_baryon(A_baryon_range)[1])):
    ax.loglog(varying_A_baryon(A_baryon_range)[0][i],
              varying_A_baryon(A_baryon_range)[1][i], color=colors[i])
    
norm_scaling = mpl.colors.Normalize(vmin=min(A_baryon_range), vmax=max(A_baryon_range))
divider = axgrid.make_axes_locatable(ax)
cax = divider.append_axes("right", size='5%', pad=0.05)
cbar = plt.colorbar(mpl.cm.ScalarMappable(norm=norm_scaling, cmap=cmap), cax=cax)
cbar.set_label("A Baryon", rotation=270, labelpad=30)

ax.set_xlabel('Wavenumber, k [h/Mpc]')
ax.set_ylabel('$P_m(k)$ [($h^{-1}Mpc)^3$]')
ax.set_title('Nonlinear Matter Power Spectra\nfor Varying A Baryon Values')
plt.show()
```
